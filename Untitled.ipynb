```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
products = pd.read_csv("Data/business.retailsales.csv")
monthly = pd.read_csv("Data/business.retailsales2.csv")

products.head()
```

```python
monthly["Month"] = monthly["Month"] + " " + monthly["Year"].astype(str)
monthly.drop("Year", inplace=True, axis=1)
monthly
```

```python
products.groupby("Product Type").count()
```

```python
# accessories = products.loc(products["Product Type"] == "Accessories")
```

```python
x_axis = monthly["Month"]
y_axis = monthly["Total Sales"]

plt.figure(figsize=(12,12))
plt.plot(x_axis, y_axis)
plt.xticks(rotation=-90)
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.show()
```

```python
x_axis2 = products["Product Type"]
y_axis2 = products["Total Net Sales"]

plt.bar(np.arange(len(x_axis2)), y_axis2, color="blue", alpha=.75, align="center")
plt.xticks(rotation=-90)
plt.title("Total Sales by Product")
plt.xlabel("Product Type")
plt.ylabel("Total Sales ($)")
plt.show()
```

```python
tick_locations = [x for x in x_axis2]
plt.xticks(tick_locations, products["Product Type"])
```
